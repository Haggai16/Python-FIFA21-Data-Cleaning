### <font color=blue> 11. Contract
**Player's contract details**

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
    return start_date, end_date, contract_length

new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i])
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)']].sample(5)
```

```python
# Contract categories

def categorize_contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'

fifa.insert(fifa.columns.get_loc('Contract Length(years)')+1, 'Contract Status', fifa['Contract'].apply(categorize_contract_status))
fifa.sample(3)
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)', 'Contract Status']].sample(5)
```

### <font color=blue> 13. Height 
**Player's jeight in centimeters**

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches* 2.54)

fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

```python
fifa = fifa.rename(columns={'Height':"Height(cm)"})
fifa.sample(1)
```

### <font color=blue> 14. Weight
**Player's weight in Kilograms

```python
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

```python
def convert_weight(weight):
    if weight is None:
        return None  # Handle None values
    if "kg" in weight:
        return int(weight.replace("kg", "").strip())  # Convert to int after stripping "kg"
    elif "lbs" in weight:
        pounds = int(weight.replace("lbs", "").strip())  # Strip "lbs" and convert to int
        return round(pounds / 2.205)  # Convert pounds to kilograms

# Apply the updated function
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
unique_weights = fifa['Weight'].unique()
print(unique_weights)
```

**<font color=green> Transforming the value, wage and release clause columns into columns of integers**

```python
fifa[['Value', 'Wage', 'Release Clause']].head()
```

```python
def convert_currency(value):
    if pd.isna(value) or isinstance(value, float):
        return value  # Return as is if NaN or already a float
    value = value.replace('€', '').strip()
    if value[-1] == 'M':
        return float(value[:-1]) * 1000000
    elif value[-1] == 'K':
        return float(value[:-1]) * 1000
    else:
        return float(value)

fifa['Value'] = fifa['Value'].apply(convert_currency)
fifa['Wage'] = fifa['Wage'].apply(convert_currency)
fifa['Release Clause'] = fifa['Release Clause'].apply(convert_currency)

fifa[['Value', 'Wage', 'Release Clause']].head()
```
